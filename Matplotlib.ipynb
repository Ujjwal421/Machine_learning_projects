```python
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
temp_data = [79,75,74,75,73,81,77]
wind_data = [14,12,10,13,9,13,12]
time_hrs = [1,2,3,4,5,6,7]
humidity_data = [73,76,78,81,81,84,84]
precipitation_data = [26,42,69,48,11,16,5]
```

```python
plt.figure(figsize=(8,4))
plt.subplots_adjust(hspace=.25)
plt.subplot(1,2,1)
plt.title('Temp')
plt.plot(time_hrs,temp_data,color = 'b',linestyle = '-',linewidth=1)
plt.subplot(1,2,2)
plt.title('Wind')
plt.plot(time_hrs,wind_data,color='r',linestyle='-',linewidth=1)
```

```python
plt.figure(figsize=(9,9))
plt.subplots_adjust(hspace=.3)
plt.subplot(2,2,1)
plt.title('Temp (F)')
plt.plot(time_hrs,temp_data,color='b',linestyle='-',linewidth=1)
plt.subplot(2,2,2)
plt.title('Wind (MPH)')
plt.plot(time_hrs,wind_data,color='r',linestyle='-',linewidth=1)
plt.subplot(2,2,3)
plt.title('Humidity (%)')
plt.plot(time_hrs,humidity_data,temp_data,color='y',linestyle='-',linewidth=1)
plt.subplot(2,2,4)
plt.title('Precipitation')
plt.plot(time_hrs,precipitation_data,color='g',linestyle='-',linewidth=1)
```

```python
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
boston_real_state_data = load_boston()
print(boston_real_state_data.DESCR)
```

```python
x_axis = boston_real_state_data.data
y_axis = boston_real_state_data.target
```

```python
style.use('ggplot')
plt.figure(figsize=(5,5))
plt.hist(y_axis,bins=50)
plt.xlabel('price in 1000s USD')
plt.ylabel('number of houses')
plt.show()
```

```python
style.use('ggplot')
plt.figure(figsize=(7,7))
plt.scatter(boston_real_state_data.data[:,5],boston_real_state_data.target)
plt.ylabel('price in 1000s')
plt.xlabel('number of houses')
plt.show()
```
